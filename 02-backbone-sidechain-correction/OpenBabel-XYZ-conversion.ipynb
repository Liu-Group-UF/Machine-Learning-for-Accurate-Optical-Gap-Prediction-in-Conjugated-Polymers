# SMILES strings export

```python
# df = pd.read_csv('./raw/sort-by-atom_num_wH-backbone-sidechain-correct-1115dp.csv')
df = pd.read_csv('./raw/atom_number_wH_sort_1115-backbone-correction-newSMILES.csv')
df.head()
```

```python
df.count()
```

```python
smiles_list = df['newSMILES']

c_smiles = []
for ds in smiles_list:
    try:
        cs = Chem.CanonSmiles(ds)
        c_smiles.append(cs)
    except:
        print('Invalid SMILES:', ds)
len(c_smiles)
```
