```python
import matplotlib.pyplot as plt
import numpy as np
```

# R2 with ECFP6 as input

```python
models = ['HBGR', 'LGBM', 'GBR', 'XGBoost', 'AdaBoost', 'RF']
r2_mfs = [0.67,0.67,0.60,0.70,0.38,0.65]
# r2_mfs_monomer = [0.71,0.71,0.64,0.72,0.46,0.66]
r2_mfs_oligomer = [0.76,0.76,0.74,0.77,0.62,0.75]

# 
bar_width = 0.25

# 
index = np.arange(len(models))
index_feature_set1 = index - bar_width*0.5
# index_feature_set2 = index
index_feature_set3 = index + bar_width*0.5

color_list=['#A39841','#67a9cf','#ef8a62']
label_list = ['ECFP6', '$E_{gap}^{monomer}$ + ECFP6', '$E_{gap}^{oligomer}$ + ECFP6']

# 
plt.figure(figsize=(12, 8), dpi=300)
plt.bar(index_feature_set1, r2_mfs, width=bar_width, label=label_list[0], color=color_list[0], alpha=0.75)
# plt.bar(index_feature_set2, r2_mfs_monomer, width=bar_width, label=label_list[1], color=color_list[1], alpha=0.75)
plt.bar(index_feature_set3, r2_mfs_oligomer, width=bar_width, label=label_list[2], color=color_list[2], alpha=0.75)

# 
# plt.xlabel('Models')
plt.ylabel('$R^2$', fontsize=40, labelpad=10)
plt.xticks(index, models)  # X
plt.ylim(0.2, 0.8)
yticks=[0.3, 0.6, 0.9]
plt.yticks(yticks)
# 
# plt.legend(fontsize=20)

# 
for spine in plt.gca().spines.values():
    spine.set_linewidth(2)

# ticks
plt.tick_params(direction='in', width=2)  # 
# plt.tick_params(axis='x', which='both', length=0, labelbottom=False)
plt.tick_params(axis='x', labelsize=30, pad=10, length=0, rotation=30)
plt.tick_params(axis='y', labelsize=24, pad=10)
# 
def add_labels(values, indexes):
    for i in range(len(values)):
        plt.text(indexes[i], values[i] + 0.01, f'{values[i]:.2f}', ha='center', fontsize=25,rotation=0) #fontsize=15

add_labels(r2_mfs, index_feature_set1)
# add_labels(r2_mfs_monomer, index_feature_set2)
add_labels(r2_mfs_oligomer, index_feature_set3)

# 
plt.show()
```
